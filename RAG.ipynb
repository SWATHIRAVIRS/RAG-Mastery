```python
from sentence_transformers import SentenceTransformer, InputExample, losses
from torch.utils.data import DataLoader

model = SentenceTransformer("BAAI/bge-large-en-v1.5")

training_set = [
    InputExample(
        texts=["patient was discharged from hospital", "hospital release summary for patient"],
        label=1.0
    ),
    InputExample(
        texts=["the stock market crashed today", "weather is very hot in Chennai"],
        label=0.0
    ),
]

data = DataLoader(training_set, shuffle=True, batch_size=16)
loss = losses.CosineSimilarityLoss(model)

model.fit(
    train_objectives=[(data, loss)],
    epochs=4,
    warmup_steps=100
)

model.save("fine-tuned-medical-report")
```

```python
import chromadb
from sentence_transformers import SentenceTransformer, util
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Step 1 - Load embedding model
model = SentenceTransformer("BAAI/bge-large-en-v1.5")

# Step 2 - Your document
document = """
Mahendra Singh Dhoni is a legendary Indian cricketer and one of the most successful
captains in the history of the sport. Born on July 7, 1981, in Ranchi, Jharkhand,
he initially worked as a ticket collector before pursuing cricket.
Known for his incredible finishing skills and calm demeanor, he earned the nickname Captain Cool.
Under his leadership, India won the 2007 ICC World Twenty20, the 2011 ICC Cricket World Cup,
and the 2013 ICC Champions Trophy.
In Test cricket, Dhoni led India to the number one ranking for the first time in 2009.
He retired from international cricket in 2020.
"""

# Step 3 - Chunk the document
splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=20
)
chunks = splitter.split_text(document)
print("Total chunks:", len(chunks))
print(chunks)

# Step 4 - Generate embeddings for all chunks
embeddings = model.encode(chunks).tolist()

# Step 5 - Create ChromaDB persistent client and collection
client = chromadb.PersistentClient(path="./my_vectordb")
collection = client.create_collection(
    name="cricket_docs",
    metadata={"hnsw:space": "cosine"}
)

# Step 6 - Store chunks in collection
collection.add(
    ids=[f"chunk{i}" for i in range(len(chunks))],
    documents=chunks,
    embeddings=embeddings,
    metadatas=[{"source": "dhoni_article", "chunk_number": i}
               for i in range(len(chunks))]
)

print("Total stored:", collection.count())

# Step 7 - Query the collection
query = "What trophies did Dhoni win?"
query_embedding = model.encode(query).tolist()

results = collection.query(
    query_embeddings=[query_embedding],
    n_results=2
)

print("\nQuery:", query)
print("\nTop results:")
for i, doc in enumerate(results['documents'][0]):
    print(f"\nChunk {i+1}:", doc)
    print("Distance:", results['distances'][0][i])
    print("Metadata:", results['metadatas'][0][i])
```

```python
from rank_bm25 import BM25Okapi
from nltk.stem import PorterStemmer

# 1. Initialize the stemmer correctly
stemmer = PorterStemmer()

# 2. Define the preprocessor to accept any list of text chunks
def preprocess_chunks(text_list):
    stemmed_chunks = []
    for text in text_list:
        # Lowercase and split into words
        words = text.lower().split()
        # Stem each word
        stemmed_word_list = [stemmer.stem(word) for word in words]
        stemmed_chunks.append(stemmed_word_list)
    return stemmed_chunks

# Test Chunks
chunks = [
    "Mahendra Singh Dhoni is a legendary Indian cricketer",
    "Under his leadership India won the 2007 ICC World Twenty20",
    "and the 2013 ICC Champions Trophy",
    "He retired from international cricket in 2020"
]

# 3. Preprocess the chunks and initialize BM25
tokenized_chunks = preprocess_chunks(chunks)
bm25 = BM25Okapi(tokenized_chunks)

# 4. Define and preprocess the query (Crucial: the query MUST be stemmed too!)
query = "What trophies did Dhoni win"
# Since preprocess_chunks expects a list, we wrap the query in a list
tokenized_query = preprocess_chunks([query])[0]

# 5. Get scores
scores = bm25.get_scores(tokenized_query)

for chunk, score in zip(chunks, scores):
    print(f"Score: {score:.4f} | {chunk}")
```

```python
import numpy as np

query_vector = np.array([0.1, 0.9, 0.05])

movie_1 = np.array([0.15, 0.85, 0.08])
movie_2 = np.array([0.5, 0.5, 0.5])
movie_3 = np.array([0.05, 0.95, 0.02])

movie_rating = np.array([movie_1, movie_2, movie_3])

dot_product = np.dot(query_vector , movie_rating)

magnitude_a = np.linalg.norm(query_vector)
magnitude_b = np.linalg.norm(movie_rating)

cosine_similarity = dot_product / (magnitude_a * magnitude_b)
print("Cosine Similarity:", cosine_similarity)
```

```python
from rank_bm25 import BM25Okapi
from sentence_transformers import SentenceTransformer, util
import numpy as np

# Setup
model = SentenceTransformer("BAAI/bge-large-en-v1.5")

chunks = [
    "signs of a cardiac arrest include chest pain",
    "heart attack rates increased globally",
    "the weather today is sunny"
]

# BM25 setup - needs tokenized (split) text
tokenized_chunks = [chunk.split() for chunk in chunks]
bm25 = BM25Okapi(tokenized_chunks)

# Semantic setup - embed all chunks once
chunk_embeddings = model.encode(chunks)

def hybrid_search(query, k=60, top_n=3):
    # BM25 scores
    tokenized_query = query.split()
    bm25_scores = bm25.get_scores(tokenized_query)
    bm25_ranks = np.argsort(-bm25_scores)  # sort descending, get indices

    # Semantic scores
    query_embedding = model.encode(query)
    semantic_scores = util.cos_sim(query_embedding, chunk_embeddings)[0].numpy()
    semantic_ranks = np.argsort(-semantic_scores)

    # Calculate RRF score for every chunk
    rrf_scores = {}

    for rank, idx in enumerate(bm25_ranks):
        rrf_scores[idx] = rrf_scores.get(idx, 0) + 1 / (k + rank + 1)

    for rank, idx in enumerate(semantic_ranks):
        rrf_scores[idx] = rrf_scores.get(idx, 0) + 1 / (k + rank + 1)

    # Sort chunks by final RRF score, highest first
    sorted_results = sorted(rrf_scores.items(), key=lambda x: x[1], reverse=True)

    # Print results
    for idx, score in sorted_results[:top_n]:
        print(f"Score: {score:.4f} | {chunks[idx]}")

# Test it
hybrid_search("heart attack symptoms")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

document = """
Marie Curie was a Polish physicist and chemist who conducted pioneering research on radioactivity.
She was the first woman to win a Nobel Prize, and the first person to win Nobel Prizes in two different sciences.
She won the Nobel Prize in Physics in 1903 along with her husband Pierre Curie.
She later won the Nobel Prize in Chemistry in 1911, this time alone, for her discovery of radium and polonium.
During World War One, she developed mobile X-ray units to help treat wounded soldiers on the battlefield.
She founded the Curie Institute in Paris, which remains a major cancer research center today.
She died in 1934 due to complications from prolonged exposure to radiation during her research.
"""

# Parent chunks - large, for context
parent_splitter = RecursiveCharacterTextSplitter(chunk_size=400, chunk_overlap=0)
parent_chunks = parent_splitter.split_text(document)

print("Number of parent chunks:", len(parent_chunks))
for i, p in enumerate(parent_chunks):
    print(f"Parent {i}: {p}\n")
```

```python
child_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=10)

child_chunks = []
child_to_parent_map = {}

child_id = 0
for parent_id, parent_text in enumerate(parent_chunks):
    children = child_splitter.split_text(parent_text)
    for child in children:
        child_chunks.append(child)
        child_to_parent_map[child_id] = parent_id
        child_id += 1

print("Number of child chunks:", len(child_chunks))
for i, c in enumerate(child_chunks):
    print(f"Child {i} (belongs to Parent {child_to_parent_map[i]}): {c}")
```

```python
from sentence_transformers import SentenceTransformer, util
import numpy as np

model = SentenceTransformer("BAAI/bge-large-en-v1.5")

# Embed only the small child chunks
child_embeddings = model.encode(child_chunks)

def parent_child_retrieve(query, top_k=1):
    query_embedding = model.encode(query)
    scores = util.cos_sim(query_embedding, child_embeddings)[0].numpy()

    best_child_idx = np.argmax(scores)
    best_parent_idx = child_to_parent_map[best_child_idx]

    print("Matched child chunk:", child_chunks[best_child_idx])
    print("\nReturning full parent context:")
    print(parent_chunks[best_parent_idx])

# Test it
parent_child_retrieve("What Nobel Prizes did Marie Curie win?")
```

```python
from sentence_transformers import CrossEncoder

reranker = CrossEncoder("BAAI/bge-reranker-base")

query = "What Nobel Prizes did Marie Curie win?"

candidates = [
    "Marie Curie was a Polish physicist and chemist who conducted pioneering research on radioactivity.",
    "She won the Nobel Prize in Physics in 1903 along with her husband Pierre Curie.",
    "During World War One, she developed mobile X-ray units to help treat wounded soldiers.",
    "She later won the Nobel Prize in Chemistry in 1911 for her discovery of radium and polonium.",
    "She founded the Curie Institute in Paris, which remains a major cancer research center today."
]

# Create query-chunk pairs
pairs = [[query, chunk] for chunk in candidates]

# Get relevance scores
scores = reranker.predict(pairs)

# Combine and sort by score
results = list(zip(candidates, scores))
results.sort(key=lambda x: x[1], reverse=True)

for chunk, score in results:
    print(f"Score: {score:.4f} | {chunk}")
```

```python
def build_rag_prompt(context_chunks, question):
    context = "\n".join([f"[{i+1}] {chunk}" for i, chunk in enumerate(context_chunks)])

    prompt = f"""
You are a helpful assistant who will answer only based on the provided context
Context: {context}
Instructions:
1. Answer the question only with the context given
2. If the question cannot be answered using the context say 'I dont know'
3. Do not use any outside knowledge beyond what is provided in the context.
4. Cite which context number(s) you used in your answer, like [1] or [2].

question: {question}
answer:
"""
    return prompt

chunks = [
    "The refund policy allows returns within 30 days of purchase.",
    "Shipping typically takes 5 to 7 business days within India.",
    "International shipping is currently not available."
]

print(build_rag_prompt(chunks, "How long does shipping take?"))
```

```python
from groq import Groq

client = Groq(api_key="your-groq-api-key")

def ask_question(question, chunks):
    prompt = build_rag_prompt(chunks, question)

    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.1
    )

    return response.choices[0].message.content
```
